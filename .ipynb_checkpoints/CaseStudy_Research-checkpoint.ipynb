## Setup

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
import textwrap

import warnings

# Code compiles so warning is unnecessary
warnings.filterwarnings("ignore", message="FixedFormatter should only be used together with FixedLocator")

fall18 = pd.read_csv("Fall 18 - Spring 19 Assessment Data.csv", encoding="Windows-1252")
fall19 = pd.read_csv("Fall 19 - Spring 20 Assessment Data.csv", encoding="Windows-1252")
fall20 = pd.read_csv("Fall 20 - Spring 21 Assessment Data.csv", encoding="Windows-1252")
fall21 = pd.read_csv("Fall 21 - Spring 22 Assessment Data.csv", encoding="Windows-1252")


# Clean data
row12 = fall18.iloc[12:13, 0:1].to_string(index=False, header=False)
row14 = fall18.iloc[14:15, 0:1].to_string(index=False, header=False)

fall18.iloc[11:12, 0:1] = fall18.iloc[11:12, 0:1] + row12
fall18.iloc[13:14, 0:1] = fall18.iloc[13:14, 0:1] + row14
fall19.iloc[11:12, 0:1] = fall18.iloc[11:12, 0:1] + row12
fall19.iloc[13:14, 0:1] = fall18.iloc[13:14, 0:1] + row14
fall20.iloc[11:12, 0:1] = fall18.iloc[11:12, 0:1] + row12
fall20.iloc[13:14, 0:1] = fall18.iloc[13:14, 0:1] + row14
fall21.iloc[11:12, 0:1] = fall18.iloc[11:12, 0:1] + row12
fall21.iloc[13:14, 0:1] = fall18.iloc[13:14, 0:1] + row14

fall18 = fall18.drop([12, 14])
fall19 = fall19.drop([12, 14])
fall20 = fall20.drop([12, 14])
fall21 = fall21.drop([12, 14])
fall19.head(26)
```

## Fall 18 - Spring 19 A1

```python
# Chart SO
df = fall18
x = 3
y = 6
z = 2
for i in range(6):
    df.iloc[x:y, 1:] = df.iloc[x:y, 1:].applymap(lambda x: pd.to_numeric(x, errors='coerce'))

    means = df.iloc[x:y, 1:].mean(axis=1)

    fig, ax = plt.subplots(figsize=(14, 8))
    ax.bar(['PI 1', 'PI 2', 'PI 3'], means, color='darkred', label='Mean Percentage', width=0.6)

    ax.axhline(y=.80, color='black', linestyle='--', label='Overall Mean')

    ax.set_xlabel('')
    ax.set_ylabel('Proportion that Met Target []Target = 80%]')
    title = ax.set_title(
        'Results by Performance Indicator associated with a Student Outcome \n' + df.iloc[z:z+1, 0:1].to_string(
            index=False, header=False))
    ax.set_ylim(0, 1)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))
    ax.yaxis.set_major_locator(mtick.MultipleLocator(0.1))
    ax.set_xticklabels([textwrap.fill(label, width=30) for label in
                        [df.iloc[x:x+1, 0:1].to_string(index=False, header=False),
                         df.iloc[x+1:x+2, 0:1].to_string(index=False, header=False),
                         df.iloc[x+2:x+3, 0:1].to_string(index=False, header=False)]], fontsize=12)

    plt.show()

    x = x + 4
    y = y + 4
    z = z + 4
```

In this analysis set. All PI's met target goal of 80%. SO 5 had particularly strong results as the means for this dataset was in the 90th percentile for all PI's.

## Fall 18 - Spring 19 A2

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import textwrap

# Read the CSV file
df = fall18

# Find the indices of the column names 'First Year', 'Sophomore', 'Junior', and 'Senior'
column_names = df.columns.tolist()
fy_start = column_names.index('First Year')
so_start = column_names.index('Sophomore')
jr_start = column_names.index('Junior')
sr_start = column_names.index('Senior')

# Define the column ranges for each class year
fy_cols = list(range(fy_start, so_start))
so_cols = list(range(so_start, jr_start))
jr_cols = list(range(jr_start, sr_start))
sr_cols = list(range(sr_start, len(column_names)))

#Start at row 2. First 2 rows are filler
current_pi_row = 3

for i in range(6):
    print('SO #', i)
    for j in range(3):
        df.iloc[current_pi_row + i, 1:] = df.iloc[current_pi_row + i, 1:].apply(lambda x: pd.to_numeric(x, errors='coerce'))
        fy_mean = df.iloc[current_pi_row + i, fy_cols].mean()
        so_mean = df.iloc[current_pi_row + i, so_cols].mean()
        jr_mean = df.iloc[current_pi_row + i, jr_cols].mean()
        sr_mean = df.iloc[current_pi_row + i, sr_cols].mean()
        fig, ax = plt.subplots(figsize=(14, 8))
        fig.text(0.1, 0.95, 'Results by Performance Indicator', fontsize=14, color='darkred')
        ax.set_title(df.iloc[current_pi_row + i, 0])
        ax.set_xlabel('')
        ax.set_ylabel('Proportion that Met Target []Target = 80%]')
        ax.bar(['First Year', 'Sophomore', 'Junior', 'Senior'], [fy_mean, so_mean, jr_mean, sr_mean], color='darkred', label='Mean Percentage', width=0.6)
        ax.axhline(y=.80, color='black', linestyle='--', label='Overall Mean')
        ax.set_ylim(0, 1)
        ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))
        ax.yaxis.set_major_locator(mtick.MultipleLocator(0.1))
        plt.show()
        
    current_pi_row += 4
        
```
