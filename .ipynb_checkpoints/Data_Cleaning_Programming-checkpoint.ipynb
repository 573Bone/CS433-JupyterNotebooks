### Part I

```python
import pandas as pd
import re

df = pd.read_csv('BL-Flickr-Images-Book.csv')

#r'\..*', '.'
#r'[\[\]]', ''

df['Title'] = df['Title'].str.replace(r'(\[.*?\]|\..*)', '', regex=True)
df['Author'] = df['Author'].str.replace(r"([\|\-].*|, (?=\w)|\..*)", '').str.title()
df['Contributors'] = df['Contributors'].str.replace(r"([\|\-].*|, (?=\w)|\..*)", '').str.title()



df.to_csv('Cleaned-BL-Flickr-Images-Book.csv', index=False)
```

### Part II

```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

df = df[df['01 !'] > 0]

df.to_csv('Cleaned_olympics.csv')

print("Top 5 gold medal earners:")
print(df.sort_values('01 !', ascending=False).head(5))

print("Countries with all three medals:")
print(df[(df['01 !.2'] > 0) & (df['02 !.2'] > 0) & (df['03 !.2'] > 0)])
```
