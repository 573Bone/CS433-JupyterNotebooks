✓ Write a custom function called cap_value(x, cap) that will set x to the cap if x > cap.
Then apply it to debt_ratio with a cap of 5.

✓ Use groupby to calculate the percent of customers that went bad for each age

✓ Add a new column age_group for the dataframe. Divide the age into groups of every 10
years in ascending order, then rearrange the data to match the age_group categorization.
Display the first eight rows of data. Do some plotting to reflect it. Use age_group for
data aggregation on revolving_utilization_of_unsecured_lines and monthly_income.
Display the result.

✓ Regroup data using number_real_estate_loans_or_lines and serious_dlqin2yrs, display
the first eight rows of data and do some plotting to reflect this it. Then do count, mean,
max, min on revolving_utilization_of_unsecured_lines and monthly_income. Display the
result. Then select the top five monthly_income values by group and display it.

```python
import pandas as pd
import matplotlib.pyplot as plt


def cap_value(x, cap):
    if x > cap:
        return cap
    else:
        return x


df = pd.read_csv("credit-data-non-null.csv")

df["debt_ratio"] = df["debt_ratio"].apply(lambda x: cap_value(x, 5))

bad_credit = df[df["serious_dlqin2yrs"] == 1].groupby("age")["serious_dlqin2yrs"].count().reset_index()
total = df.groupby("age")["serious_dlqin2yrs"].count().reset_index()

output = pd.merge(total, bad_credit, on="age", how="left")
output["percent_bad_customers"] = output["serious_dlqin2yrs_y"] / output["serious_dlqin2yrs_x"] * 100
output["percent_bad_customers"] = output["percent_bad_customers"].fillna(0)

print(output[["age", "percent_bad_customers"]].to_string(index=False))


df['age_group'] = pd.cut(df['age'], bins=range(0, 110, 10), right=False)
df = df.sort_values(by='age_group')

print(df.head(8).to_string())

df['age_group'].value_counts().sort_index().plot(kind='bar', xlabel='Ages', ylabel='Number of Customers')
plt.show()

data = df.groupby('age_group')[['revolving_utilization_of_unsecured_lines', 'monthly_income']].agg(['mean', 'median'])

print(data)


real_estate_and_dlq = df.groupby(["number_real_estate_loans_or_lines", "serious_dlqin2yrs"])
print(real_estate_and_dlq.head(8))

real_estate_and_dlq.size().unstack().plot(kind='bar', xlabel='Loans', ylabel='Income')
plt.show()

output = real_estate_and_dlq.agg({"revolving_utilization_of_unsecured_lines": ["count", "mean", "max", "min"],
                                  "monthly_income": ["count", "mean", "max", "min"]})
print(output)

print(real_estate_and_dlq['monthly_income'].nlargest(n=5).to_string())
```
